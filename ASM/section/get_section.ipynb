```python
import numpy as np
import sys
```

```python
data_path = "E:/train/train/"
to_save = "../../Dataset/data/section/"
c_files = []
files_not_parsed = []
dic_section = []
```

```python
malwarelist = np.genfromtxt('../../Dataset/trainLabels.csv', delimiter=",", dtype=str)
malwarelist = malwarelist[1:]
```

```python
count = 0
for name in malwarelist:
    head = {}
    nam = name[0].strip('"')
    if nam not in c_files:
        if nam not in files_not_parsed:
            nam = str(nam)
            nam_asm = data_path + nam + ".asm"
            nam_section = to_save + nam + ".txt"
            try:
                with open(nam_asm, encoding='ISO-8859-1') as f:
                    lines = f.readlines()
                    for source_line in lines:
                        if source_line == "\n":
                            continue
                        xx = source_line.split()
                        h = xx[0].split(':')
                        if h[0] not in dic_section:
                            dic_section.append(h[0])
                        if h[0] not in head:
                            head[h[0]] = 0
                        head[h[0]] += 1
                with open(nam_section, 'w', encoding='ISO-8859-1') as file_section:
                    for i in head:
                        file_section.write(str(i) + ' ' + str(head[i]) + '\n')
                c_files.append(nam)
            except:
                files_not_parsed.append(nam)
    print("finish file ", count + 1)
    count += 1
np.save('../store/section/c_file.npy', c_files)
np.save('../store/section/files_not_parsed.npy', files_not_parsed)

np.save('../store/section/dic_section.npy',dic_section)
```
