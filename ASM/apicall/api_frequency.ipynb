```python
import numpy as np
import sys
import os
import csv
```

```python
path = "../Dataset/data/apicall/"
```

```python
malwarelist=np.genfromtxt('../Dataset/trainLabels.csv',delimiter=",",dtype=str)
malwarelist=malwarelist[1:]
```

```python
dic_api = np.load('./store/api/dic_api.npy')
dic_api = dic_api.tolist()
dic_api.append('ID')
dic_api.append('LABELS')  

wrong = np.load('./store/api/files_not_parsed.npy')
wrong = wrong.tolist()
data_to_write = []
data_to_write.append(dic_api)
```

```python
print(wrong)
```

```python
#print(len(dic_api))
for i in range(1, 5):
    print(i)
    sys.exit()
```

```python
count = 0
for name in malwarelist:
    nam = name[0].strip('"')
    nam = str(nam)
    count = count + 1
    print(nam, ' ', count)
    temp = [0] * len(dic_api)
    nam = path + nam + ".txt"
    if name[0].strip('"') not in wrong:
        with open(nam) as f:
            lines = f.readlines()
            for source_line in lines:
                if source_line == '\n':
                    continue
                source_line = source_line.strip('\n')
                if source_line in dic_api:
                    temp[dic_api.index(source_line)] = temp[dic_api.index(source_line)] + 1
            temp[dic_api.index('ID')] = name[0].strip('"')
            temp[dic_api.index('LABELS')] = name[1]
    data_to_write.append(temp)
```

```python
myFile = open('../Dataset/data/data_apicall/data_apicall.csv','w', newline = '')
with myFile:
    write = csv.writer(myFile)
    write.writerows(data_to_write)
```
