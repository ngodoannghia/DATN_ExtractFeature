```python
import numpy as np
import string
import csv
import sys
path = '../../Dataset/data/1gram/'
```

```python
malwarelist=np.genfromtxt('../../Dataset/trainLabels.csv',delimiter=",",dtype=str)
malwarelist=malwarelist[1:]
```

```python
dic_opcode = np.load('../store/ngram/dic_1gram.npy')
dic_opcode = dic_opcode.tolist()
dic_opcode.append('LABELS')
dic_opcode.append('ID')
```

```python
data_to_write = []
data_to_write.append(dic_opcode)
```

```python
count = 0

for name in malwarelist:
    nam = name[0].strip('"')
    nam_txt = path + nam +'.txt'
    temp = [0] * len(dic_opcode)
    print(nam, ' ', count)
    with open(nam_txt, 'r') as f:
        lines = f.read()
        lines = lines.split('\n')
        for source_line in lines:
            if source_line in dic_opcode:
                temp[dic_opcode.index(source_line)] = temp[dic_opcode.index(source_line)] + 1
        temp[dic_opcode.index('ID')] = name[1]
        temp[dic_opcode.index('LABELS')] = name[0].strip('"')
    data_to_write.append(temp)
    count += 1
```

```python
myFile = open('../../Dataset/data/data_ngram/data_1gram.csv','w', newline = '')
with myFile:
    write = csv.writer(myFile)
    write.writerows(data_to_write)
```
